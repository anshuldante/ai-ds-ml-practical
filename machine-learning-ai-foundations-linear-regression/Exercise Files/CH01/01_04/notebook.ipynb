```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
from sklearn.metrics import r2_score
from mpl_toolkits import mplot3d
import math
```

```python
df = pd.read_csv('../../mydata/auto-imports.csv', header=None, na_values='?')
num_map = {'two':2, 'four':4, 'six':6, 'five':5, 'three':3, 'twelve':12, 'two':2, 'eight':8}

df.columns = ['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels',
             'engine-location','wheel-base','length','width','height','curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower','peak-rpm','city-mpg','highway-mpg','price']

df['cylinders'] = df['num-of-cylinders'].map(num_map)
df['doors'] = df['num-of-doors'].map(num_map)
```

```python
df = df.dropna(subset=['price'])
df = df.reset_index(drop=True)
df
```

```python
df.horsepower=df.horsepower.fillna(df.horsepower.mean())
df.horsepower.unique()
```

```python
df.columns,df['horsepower'].unique(), df['engine-size'].unique(), df['price'].unique()
```

```python
colorMap = {"sedan": "red","hatchback": "green","wagon": "blue","hardtop": "yellow","convertible": "black"}
colors = df['body-style'].map(colorMap)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(df['horsepower'], df['engine-size'], df['price'],c=colors, marker='.')
ax.set_xlabel('Horsepower')
ax.set_ylabel('Engine Size')
ax.set_zlabel('Price')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.scatter(df['horsepower'],df['price'],c=colors, marker='.')
ax.set_xlabel('Horsepower')
ax.set_ylabel('Price')

#for index in range(0, len(df.index)):
#    ax.annotate(df.iloc[index]['make'], (df.iloc[index]['horsepower'],df.iloc[index]['price']))
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.scatter(df['horsepower'],df['price'],c=colors, marker='.', s=df['engine-size'])
ax.set_xlabel('Horsepower')
ax.set_ylabel('Price')

#for index in range(0, len(df.index)):
#    ax.annotate(df.iloc[index]['make'], (df.iloc[index]['horsepower'],df.iloc[index]['price']))

z = np.polyfit(df['horsepower'], df['price'], 1)
f = np.poly1d(z)
print(f"{f} - fit line\n{r2_score(df['horsepower'], f(df['price']))} - R2")
ax.plot(df['horsepower'], f(df['horsepower']), color='black')


plt.show()
```
