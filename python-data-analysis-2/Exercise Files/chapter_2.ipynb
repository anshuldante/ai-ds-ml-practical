```python
# Find all possible combinations of coins (1,2,5,10,25,50,100) and they should sum up to X INR (X*100 paise)
# Brute-force

def getRange(denomination, X = 1):
    return int((X * 100 / denomination) + 1)
```

```python
print(getRange(1))
print(getRange(2))
print(getRange(5))

print(getRange(1, 3))
print(getRange(2, 3))
print(getRange(5, 3))
```

```python
X = 1
combinations = []

for count_100 in range(getRange(100, X)):
    for count_50 in range(getRange(50, X)):
        for count_25 in range(getRange(25, X)):
            for count_10 in range(getRange(10, X)):
                for count_5 in range(getRange(5, X)):
                    for count_2 in range(getRange(2, X)):
                        for count_1 in range(getRange(1, X)):
                            if (100 * count_100 + 50 * count_50 + 25 * count_25 + 10 * count_10 + 5 * count_5 + 2 * count_2 + 1 * count_1 == X * 100):
                                combinations.append([count_100, count_50, count_25, count_10, count_5, count_2, count_1])
```

```python
# same as above, but now as a function that takes arbitrary total

def find_combinations(total):
    combinations_amounts = []

    for amount_100 in range(0, total+1, 100):
        for amount_50 in range(0, total+1, 50):
            for amount_25 in range(0, total+1, 25):
                for amount_10 in range(0, total+1, 10):
                    for amount_5 in range(0, total+1, 5):
                        total_so_far = amount_100 + amount_50 + amount_25 + amount_10 + amount_5

                        if total_so_far <= total:
                            combinations_amounts.append([amount_100, amount_50, amount_25, amount_10, amount_5,
                                                         total - total_so_far])
    
    return combinations_amounts
```

```python
totals = range(100, 1100, 100)
lengths = [len(find_combinations(total)) for total in totals]
```
