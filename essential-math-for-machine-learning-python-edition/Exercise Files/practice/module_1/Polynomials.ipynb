```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# 3ùë•3‚àí4ùë•+5
```

```python
df = pd.DataFrame({'x' : range(1,50)})
```

```python
df['y'] = 3 * (df['x'] ** 3) - 4 * df['x'] + 5
```

```python
plt.plot(df['x'], df['y'], 'blue')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline()
plt.axvline()
plt.show()
```
