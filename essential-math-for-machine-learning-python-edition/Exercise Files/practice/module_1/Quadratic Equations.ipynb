```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# 𝑦=2𝑥2+2𝑥−4

df = pd.DataFrame({'x' : range ( -20,20)})
df['y'] = 2 * (df['x'] ** 2) + 2 * df['x'] - 4

plt.plot(df['x'], df['y'], 'green')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline()
plt.axvline()

plt.plot()
```

```python
# 𝑦=−2𝑥2+6𝑥+7

df = pd.DataFrame({'x' : range ( -20,+24)})
df['y'] = - 2 * (df['x'] ** 2) + 6 * df['x'] + 7

plt.plot(df['x'], df['y'], 'red')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline()
plt.axvline()

plt.plot()
```
