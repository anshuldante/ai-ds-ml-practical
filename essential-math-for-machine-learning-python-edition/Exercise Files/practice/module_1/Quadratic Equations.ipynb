```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
# ùë¶=2ùë•2+2ùë•‚àí4

df = pd.DataFrame({'x' : range ( -20,20)})
df['y'] = 2 * (df['x'] ** 2) + 2 * df['x'] - 4

plt.plot(df['x'], df['y'], 'green')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline()
plt.axvline()

plt.plot()
```

```python
# ùë¶=‚àí2ùë•2+6ùë•+7

df = pd.DataFrame({'x' : range ( -20,+24)})
df['y'] = - 2 * (df['x'] ** 2) + 6 * df['x'] + 7

plt.plot(df['x'], df['y'], 'red')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline()
plt.axvline()

plt.plot()
```
