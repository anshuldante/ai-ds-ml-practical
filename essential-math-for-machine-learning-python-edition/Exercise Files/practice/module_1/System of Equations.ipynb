```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# x + y = 16
# 10x + 25 y = 250

chip10s = [0, 16]
chip25s = [16, 0]

chip10values = [25, 0]
chip25values = [0, 10]
```

```python
# Plot the lines
plt.plot(chip10s,chip25s, color='blue')
plt.plot(chip10values, chip25values, color="orange")
plt.xlabel('x (£10 chips)')
plt.ylabel('y (£25 chips)')
plt.grid()

plt.show()
```

```python
# apples + oranges = 9
# 5apval + 15orval = 75

apples = [0, 9]
oranges = [9, 0]

apval = [15, 0]
orval = [0, 5]

plt.plot(apples, oranges, color = 'green')
plt.plot(apval, orval, color = 'orange')

plt.xlabel('Apples')
plt.ylabel('Oranges')
plt.grid()

plt.show()
```
