```python
import pandas as pd
```

```python
import glob
```

```python
# get data file names
path = r'Exercise Files'
filenames = glob.glob(path + "/crime_boston_*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
crime = pd.concat(dfs, ignore_index=True)
```

```python
crime.isnull()
```

```python
crime.isnull().any(axis = 1)
```

```python
rows_with_missing_vals = crime.isnull().any(axis = 1)
crime[rows_with_missing_vals]
```
